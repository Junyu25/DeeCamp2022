```python
features = KOlist[0:12]
```

```python
KOlist = ['GLCabcpp', 'GLCptspp', 'HEX1', 'PGI', 'PFK', 'FBA', 'TPI', 'GAPD', 'PGK', 'PGM', 'ENO', 'PYK']
```

```python
x = df.loc[:, features].values
```

```python
y = df.loc[:,['comb']].values
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
```

```python
principalDf 
```

```python
finalDf = pd.concat([principalDf, df[['comb']]], axis = 1)
```

```python
finalDf
```

```python
df1 = pd.merge(df, finalDf, on="comb")
```

```python
df1 = df1[df1['obj'] > 2]
```

```python
df1["Production"] = df1["obj_rand"]
```

```python
df1
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(df1, x='PC1', y='PC2', z='obj',
              color='obj', width=800, height=650)
fig.show()
```
