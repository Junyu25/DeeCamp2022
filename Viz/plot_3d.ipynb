```python
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../GEMs/Ecoli_Suc_KO_10_PCA.csv")
```

```python
df
```

```python
features = ['GAPD', 'PGK', 'PGM', 'ENO', 'TPI', 'ACONTb', 'ACONTa', 'CS', 'FUM', 'MDH']
```

```python
x = df.loc[:, features].values
```

```python
y = df.loc[:,['comb']].values
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf 
```

```python
finalDf = pd.concat([principalDf, df[['comb']]], axis = 1)
```

```python
finalDf
```

```python
df1 = pd.merge(df, finalDf, on="comb")
```

```python
df1
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(df1, x='principal component 1', y='principal component 2', z='obj',
              color='obj')
fig.show()
```

How can you conect the scatter plot? or conect the dots to the line -> surface of landscape?

```python
import plotly.graph_objects as go
```

```python
z = df1.obj.values
x, y = list(df1["principal component 1"]), list(df1["principal component 2"])
```

```python
type(x)
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D fitness landscape', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```
